```python
# time resampling
data.head()
```

```python
# year and freq
data.resample(rule='A').max()['Open'].plot()
```

```python
# quaterly start frequency
data.resample(rule='QS').max()['High'].plot()
```

```python
# Business end frequency
data.resample(rule='BA').max()['High'].plot()
```

```python
# Business quat end frequency
data.resample(rule='BQS').max()['High'].plot()
```

```python
# plotting
data['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
# plotting
data['Open'].resample(rule='M').max().plot(kind='bar', figsize=(35,6))
```

```python
data['High'].rolling(10).max().head(20)
```

```python
data['Open:30 days rolling'] = data['Open'].rolling(30).mean()
```

```python
data.head(31)
```

```python
data[['Open', 'Open:30 days rolling']].plot(figsize=(12,5))
```
